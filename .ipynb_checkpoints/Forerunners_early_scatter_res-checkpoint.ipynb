```python
# Transition probabilities matrix
def transition_matrix(transitions):
    n = 1+ max(transitions) #number of states

    M = [[0]*n for _ in range(n)]

    for (i,j) in zip(transitions,transitions[1:]):
        M[i][j] += 1

#convert to probabilities:
    for row in M:
        s = sum(row)
        if s > 0:
            row[:] = [f/s for f in row]
    return M
```

```python
unique_block_numbers = df_groupby.block_number.unique()
n_clusters = kmeans.n_clusters
```

```python
# Transition probability
data = []
for block_number in unique_block_numbers:
    df_hui = df_groupby.loc[df_groupby.block_number == block_number]
    t = df_hui.cluster_rank
    m = transition_matrix(t)
    data.append(m)

results_dic = {}
for x in range(0, n_clusters):
    for y in range(0, n_clusters):
        step = []
        for i in range(0, len(data)):
            step.append(data[i][x][y])
        results_dic[str(x) + str(y)] = step
```

```python
data = []
for block_number in unique_block_numbers:
    df_h = subj_1.loc[subj_1.block_number == block_number]
    t = df_h.cluster_rank
    m = transition_matrix(t)
    data.append(m)

results_dic = {}
for x in range(0,4):
    for y in range(0,4):
        step = []
        for i in range(0, len(data)):
            step.append(data[i][x][y])
        results_dic[str(x) + str(y)] = step
```

```python
#Create df of transition
results_0= pd.DataFrame.from_dict(results_dic)
results_0.reset_index(level=0, inplace=True)
```

```python
unique_transition = list(results_0.columns) 
unique_transition.remove('index')
```

```python
#Linear regres coef 
from scipy import stats
reg_coef = []
for transition in unique_transition:
    a =[]
    x = results_0.index
    y = results_0[transition]
    a.append(transition)
    slope, intercept, r_value, p_value, std_err = stats.linregress(x,y)
    a.append(slope)
    a.append(intercept)
    a.append(r_value) 
    a.append(p_value) 
    a.append(std_err) 
    reg_coef.append(a) 
```

```python
#Linear regres coef dataframe
reg_coef_df = pd.DataFrame(reg_coef) 
#pd.set_option('display.max_rows', 1000)
reg_coef_df = reg_coef_df.rename(columns={0: 'transition', 1: "slope", 2: "intercept", 3: "r_value", 4: "p_value", 5: "std_err"})
reg_coef_df.reset_index(drop=True, inplace=True)
```

```python
# p_value < 0.05 
reg_coef_df.loc[reg_coef_df['p_value'] < 0.05]
```

```python
#FDR_correction
import statsmodels.sandbox.stats.multicomp as multicomp
pval_corrected = multicomp.multipletests(reg_coef_df['p_value'], method='fdr_bh', is_sorted=False, returnsorted=False)[1]
reg_coef_df['pval_corrected'] = list(pval_corrected)
```

```python
# p_value_corec < 0.05 
reg_coef_df.loc[reg_coef_df['pval_corrected'] < 0.06]
```
