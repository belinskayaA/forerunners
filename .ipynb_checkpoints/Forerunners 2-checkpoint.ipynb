```python
def band_hilbert(x, fs, band, N=None, axis=-1):
    x = np.asarray(x)
    Xf = np.fft.fft(x, N, axis=axis)
    w = np.fft.fftfreq(x.shape[0], d=1. / fs)
    Xf[(w < band[0]) | (w > band[1])] = 0
    x = np.fft.ifft(Xf, axis=axis)
    return 2*x
```

### The clustering of the 3D vectors - estimates of the instantaneous power of 𝛂, 𝛃 and, 𝜽 rhythms. 

```python
# Optimal number of clusters
Sum_of_squared_distances = []
K = range(1,25)
for k in K:
    km = KMeans(n_clusters=k)
    km = km.fit(df_groupby[['norm_env_beta','norm_env_alpha','norm_env_theta']])
    Sum_of_squared_distances.append(km.inertia_) 
plt.plot(K, Sum_of_squared_distances, 'bx-')
plt.xlabel('Number_of_clusters')
plt.ylabel('Sum_of_squared_distances')
plt.show() 
```

### Construction of the transition matrix from cluster to cluster.

```python
#Sort all clusters from min to max mean env_alpha
clusters = np.arange(kmeans.n_clusters)
alpha_values = kmeans.cluster_centers_[:,1]
  
# create dict wich map original cluster number to cluster alpha rank
cluster_to_rank_dict = dict(zip(clusters[np.argsort(alpha_values)], np.arange(len(clusters))))
cluster_to_rank_dict
```

```python
#how to map clusters to clusters runk
df_groupby['cluster_rank'] = df_groupby['cluster'].map(cluster_to_rank_dict)
```

```python
# Split up for 15 blocks
df_groupby['block_number'] = df_groupby.index.map(lambda a : int(a / (len(df_groupby) / 15)))
```

```python
# Transition probabilities matrix
def transition_matrix(transitions):
    n = 1+ max(transitions) #number of states

    M = [[0]*n for _ in range(n)]

    for (i,j) in zip(transitions,transitions[1:]):
        M[i][j] += 1

#convert to probabilities:
    for row in M:
        s = sum(row)
        if s > 0:
            row[:] = [f/s for f in row]
    return M
```

```python
unique_block_numbers = df_groupby.block_number.unique()
n_clusters = kmeans.n_clusters - 1
```

```python
# Transition probability
data = []
for block_number in unique_block_numbers:
    df_h = df_groupby.loc[df_groupby.block_number == block_number]
    t = df_h.cluster_rank
    m = transition_matrix(t)
    data.append(m)

results_dic = {}
for x in range(0,n_clusters):
    for y in range(0,n_clusters):
        step = []
        for i in range(0, len(data)):
            step.append(data[i][x][y])
        results_dic[str(x) + str(y)] = step
```

```python
#Create df of transition
results_0= pd.DataFrame.from_dict(results_dic)
results_0.reset_index(level=0, inplace=True)
```

## We used linear regression to see how the probability of each transition changes as a function of the training segment.

```python
#Linear regres coef 
from scipy import stats
unique_transition = list(results_0.columns) 
unique_transition.remove('index')
reg_coef = []
for transition in unique_transition:
    a =[]
    x = results_0.index
    y = results_0[transition]
    a.append(transition)
    slope, intercept, r_value, p_value, std_err = stats.linregress(x,y)
    a.append(slope)
    a.append(intercept)
    a.append(r_value) 
    a.append(p_value) 
    a.append(std_err) 
    reg_coef.append(a) 
```

```python
#Linear regres coef dataframe
reg_coef_df = pd.DataFrame(reg_coef) 
#pd.set_option('display.max_rows', 1000)
reg_coef_df = reg_coef_df.rename(columns={0: 'transition', 1: "slope", 2: "intercept", 3: "r_value", 4: "p_value", 5: "std_err"})
reg_coef_df.reset_index(drop=True, inplace=True)
```

```python
# p_value < 0.05 
reg_coef_df.loc[reg_coef_df['p_value'] < 0.05]
```

### FDR_correction

```python
import statsmodels.sandbox.stats.multicomp as multicomp
pval_corrected = multicomp.multipletests(reg_coef_df['p_value'], method='fdr_bh', alpha=0.3, is_sorted=False, returnsorted=False)[1]
reg_coef_df['pval_corrected'] = list(pval_corrected)
```
